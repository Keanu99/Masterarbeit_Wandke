```python
#Nach jeder Messung ausführen und anpassen der Werte zum Loggen 

#Auswertung Power / Leistungsaufnahme. Wird für die spätere Auswertung benötgit. Daten werden zusätzlich geloggt.
#Die folgenden Auswertung muss pro Messungen durchgeführt werden. 
#Wichtig ist hier die korrekte Angabe der unten stehenden Attribute: datensatz_name, frequency und voltage, die passend zur auszuwertenden Messung gehören. 
#Die durchschnittliche Leistungsaufnahme der CPU wird hier in einer extra Datei geloggt. Damit diese später gesamt ausgewertet werden können.


import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook

# CSV-Datei einlesen
daten = pd.read_csv(csv_datei)

# Berechnung des Energieverbrauchs
letzte_zeile_index = daten.index[-1]
durchschnitt_power = daten['CPU-Gesamt-Leistungsaufnahme [W]'].mean()
letzter_wert_steps = daten.iloc[letzte_zeile_index]['Zeit']
gesamt_zeit = (letzter_wert_steps * 100) / 1000
energieverbrauch = durchschnitt_power * letzter_wert_steps
plt.figure()
# Liniendiagramm erstellen und Durchschnitt als Linie hinzufügen
plt.plot(daten['Zeit'], daten['CPU-Gesamt-Leistungsaufnahme [W]'], label='CPU-Leistungsaufnahme')
plt.axhline(y=durchschnitt_power, color='r', linestyle='--', label=f'Durchschnitt = {durchschnitt_power:.2f} W')

plt.plot(daten['Zeit'], daten['CPU Kern-Leistung (SVI2 TFN) [W]'], label='CPU-Kern-Leistungsaufnahme')
durchschnitt_power_core = daten['CPU Kern-Leistung (SVI2 TFN) [W]'].mean()
plt.axhline(y=durchschnitt_power_core, color='b', linestyle='--', label=f'Durchschnitt = {durchschnitt_power_core:.2f} W')

plt.xlabel('Zeit in Sekunden')
plt.ylabel('Leistungsaufnahme in Watt')
plt.grid(True)
#plt.ylim(bottom=0)
#plt.ylim(top=5000)
plt.legend()
plt.show()

from datetime import datetime

def time_diff_in_seconds(time1, time2):
    """
    Berechnet die Differenz in Sekunden zwischen zwei Zeitstempeln.
    """
    fmt = '%H:%M:%S.%f'
    t1 = datetime.strptime(time1, fmt)
    t2 = datetime.strptime(time2, fmt)
    diff = t2 - t1
    return diff.total_seconds()

def integrate_data(data):
    total_integral = 0
    for i in range(len(data) - 1):
        time1 = data['Time'][i]
        time2 = data['Time'][i+1]
        value1 = data['CPU-Gesamt-Leistungsaufnahme [W]'][i]
        value2 = data['CPU-Gesamt-Leistungsaufnahme [W]'][i+1]
        time_diff = time_diff_in_seconds(time1, time2)
        integral = time_diff * ((value1 + value2) / 2) 
        total_integral += integral
    return total_integral

energieverbrauch2 = integrate_data(daten)
#print("Nicht integriert: ", energieverbrauch)
#print("Integriert: ", energieverbrauch2)
#Zum loggen der durchschnittlichen Leistungsaufnahmen des jeweiligen Messvorgangs
import csv
import os

def integrate_data_gpu(data):
    total_integral = 0
    for i in range(len(data) - 1):
        time1 = data['Time'][i]
        time2 = data['Time'][i+1]
        value1 = data['GPU Leistung [W]'][i]
        value2 = data['GPU Leistung [W]'][i+1]
        time_diff = time_diff_in_seconds(time1, time2)
        integral = time_diff * ((value1 + value2) / 2) 
        total_integral += integral
    return total_integral

energieverbrauch_GPU = integrate_data_gpu(daten)

def add_row_to_csv(filename, datensatz_name, frequency, voltage, energieverbrauch, avg_power, new_value):
    # Daten für die neue Zeile
    energieverbrauch_gpu_cpu = round(energieverbrauch + new_value, 2)
    cpu_anteil = round((100 / energieverbrauch_gpu_cpu) * energieverbrauch, 2)
    new_row = {'Datensatz': datensatz_name, 'Frequenz': frequency, 'voltage': voltage, 'Energieverbrauch': energieverbrauch, 'avg_power': avg_power, 'GPU-Energieverbrauch': new_value, 'Energieverbrauch_gpu_cpu': energieverbrauch_gpu_cpu, 'Cpu-Anteil-Energieverbrauch': cpu_anteil}
    
    # Überprüfe, ob die Datei existiert
    file_exists = os.path.isfile(filename)
    
    # Öffne die CSV-Datei im Modus 'a' (append), um Werte hinzuzufügen
    with open(filename, 'a', newline='') as csvfile:
        # Erstelle einen CSV-Writer, wenn die Datei neu erstellt wird
        writer = csv.DictWriter(csvfile, fieldnames=['Datensatz', 'Frequenz', 'voltage', 'Energieverbrauch', 'avg_power', 'GPU-Energieverbrauch', 'Energieverbrauch_gpu_cpu', 'Cpu-Anteil-Energieverbrauch'])
        
        # Schreibe die Kopfzeile, falls die Datei neu erstellt wurde
        if not file_exists:
            writer.writeheader()
        
        # Schreibe die neue Zeile in die CSV-Datei
        writer.writerow(new_row)

# Beispielaufruf der Funktion
filename = 'test_spannung.csv'

datensatz_name = 'cifar'
frequency = '2000'
voltage = '1.1'
avg_core_power = round(durchschnitt_power_core, 2)
avg_power = round(durchschnitt_power, 2)
energieverbrauch_integriert = round(energieverbrauch2,2)
new_value = round(energieverbrauch_GPU, 2)


add_row_to_csv(filename, datensatz_name, frequency, voltage, energieverbrauch_integriert, avg_power, new_value)


#plt.savefig(r'C:\Users\keanu\OneDrive\Desktop\Master\Masterarbeit\Versuche\ResNet_(Spannung)\Cifar\power_ResNet_Cifar_3600_1,1.png', dpi=300)
```
